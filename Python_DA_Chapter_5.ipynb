# TXT

TXT files can containg separators that are not well defined, for this cases we can use de RegExp.

See the table of metacharacters:


* .  single character, except newline
* \d digit
* \D non-digit character
* \s whitespace character
* \S non-whitespace character
* \n new line character
* \t tab character
* \uxxxx unicode character specified by the hexadecimal number xxxx



```python
import re #to import regexp
import pandas as pd
```

```python
%%writefile ch05_04.txt
white red blue green
1 5 2 3
2 7 8 5
3 3 6 7
```

```python
pd.read_table('ch05_04.txt', sep='\s')
```

```python
%%writefile ch05_05.txt
000END123AAA122
001END124BBB321
002END125CCC333
```

```python
#header option set to None whenever the column headings are not present in the TXT file
#for this is used 'keep_default_na= false' or 'na_filter=False' to values 'NaN' not be shown
pd.read_table('ch05_05.txt', sep='\D', header=None, na_filter=False) 
```

```python
%%writefile ch05_06.txt
########### LOG FILE ############
This file has been generated by automatic system
white,red,blue,green,animal
12-Feb-2015: Counting of animals inside the house
1,5,2,3,cat
2,7,8,5,dog
13-Feb-2015: Counting of animals outside the house
3,3,6,7,horse
2,2,8,3,duck
4,4,2,1,mouse
```

```python
pd.read_table('ch05_06.txt', sep=',', skiprows=[0,1,3,6])
```

```python
#With 'nrows' and 'skiprows' options, you can select the starting line n (n = SkipRows) and the lines to be read after it (nrows = i)

pd.read_csv('myCSV_02.csv', skiprows=[2], nrows=3, header=None)
```

```python
out = pd.Series()
i = 0
pieces = pd.read_csv('myCSV_01.csv', chunksize=3)

for piece in pieces:
  out.at[i, piece['white'].sum()] #From the book this line is: out.set_value(i,piece['white'].sum()) || "set_value" is Deprecated since version 0.21.0: Use .at[] or .iat[] accessors instead.
  i += 1
```
